```python
import re
import numpy as np
import pandas as pd
import xlwt
import ete3
```

```python
df = pd.read_excel("../../Databases/PHISTO/Shigella flexneri.xls", names = ["Pathogen","Taxid_Pathogen","Pathogen_protein", "Human_protein", "Detection_method", "Pubmed_ID"], usecols= "A,B,G,I:L,N")
```

```python
df=df.drop("Pathogen Protein", axis=1)
df=df.drop("Human Protein", axis=1)
```

```python
df.columns=["Pathogen","Taxid_Pathogen","Pathogen_protein", "Human_protein", "Detection_method", "Pubmed_ID"]
df=df.sort_values("Pathogen_protein")

ncbi = ete3.NCBITaxa()

test=df["Taxid_Pathogen"]
test = ncbi.translate_to_names(test)
df["Pathogen"]=test
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_excel("../Test outputs/Shigella flexneri PHISTO.xls")
```
